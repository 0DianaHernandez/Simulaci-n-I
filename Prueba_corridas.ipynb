**Pueba de corridas**

Recuerde que las dos propiedades más importantes que deben satisfacer los números de un conjunto x_i, son uniformidad e independencia.

Esta es una prueba estadística que trata de corroborar si los números en el intervalo (0, 1) son independientes o, en otras palabras, si parecen pseudoaleatorios.
Para probar la independencia de los números de un conjunto x_i primero es preciso formular las siguientes hipótesis:

*   H_0: Los números del conjunto x_i, son independientes
*   H_1: Los números del conjunto x_i, no son independientes

*Procedimiento*

1. Determinar una secuencia de números (S) que sólo contiene unos y ceros, de acuerdo con una comparación entre x_i y x_i-1:

                          * 0 si x_i<= x_i-1
                          * En caso contrario, se escribe 1


2. Después se determina el número de corridas observadas, C_0 (una corrida se identifica como la cantidad de unos o ceros consecutivos).
3. Luego se calcula el valor esperado, la varianza del número de corridas y el estadístico Z_0 mediante las ecuaciones:

                          * μ_C_0= (2n-1)/3

                          * σ^2_C_0= (16n-29)/90

                          * Z_0= | (C_0 - μ_C_0)/σ_C_0 |

4. Evalución de las hipótesis: Buscar en la tabla de valores de una distribución estandar el calor correspondiente σ/2,

* Si el estadístico Z_0 es mayor que el valor crítico de -Z_α/2, y menor o igual que Z_α/2, se concluye que los números del conjunto x_i no son independientes.
* De lo contrario no se puede rechazar que el conjunto de x_i sea independiente.


```python
'''EJERCICIO1: Realice una prueba de corridas con un nivel de aceptación del 95 %'''
from tabulate import tabulate
from scipy.stats import norm
import math

# Definimos la lista de Valores
lista= [0.048, 0.591, 0.478, 0.166, 0.932,0.926, 0.007, 0.551, 0.362, 0.533,0.368, 0.437, 0.178, 0.411, 0.545,0.787, 0.594, 0.086, 0.951, 0.772,0.298, 0.199, 0.157, 0.401, 0.631]

#1. Definimos nuestras hipotesis
H_0=["lOS NÚMEROS DE LA MUESTRA SON INDEPENDIENTES "]
H_1=["LOS NÚMEROS DE LA MUESTRA NO SON INDEPENDIENTES "]


# 2. Definimos el tamaño de "n" y el nivel de significancia α
n=len(lista)
Nivel_aceptacion=int(input("Ingrese el nivel de significancia:"))
Nivel_significancia= Nivel_aceptacion/100
alpha= round(1- Nivel_significancia,2)

print("N es:",n)
print("Nivel de significancia:",Nivel_significancia)
print("α es:",alpha)


# 3. Definimos la secuencia S mediante un ciclo for
Secuencias=[]

for i in range(n):
  x_i=lista[i]
  x_i_1=lista[i-1]
  if i==0:
    secuencias="-" #Esto solo es porque el primer elemento de la lista no lo contaremos sino hasta el segundo que es donde se comprara si aumenta o disminuye el valor
  elif x_i > x_i_1:
     secuencias=1 #Si aumenta el valor de x_i+1 (x_i) con respecto a x_i (x_i_1) le asignamos 1
  elif x_i < x_i_1: #Si disminuye el valor de x_i+1 con respecto a x_i le asignamos 0
    secuencias=0
# Agregamos los resultados de esta iteración a la lista
  Secuencias.append([i, x_i, secuencias])

# Encabezados de la tabla
headers = ["Iteración", "x_i", "Secuencia"]
# Imprimir la tabla usando la función tabulate
tabla = tabulate(Secuencias, headers=headers, tablefmt="grid")
print(tabla)


# 4.Definimos la segunda tabla donde se mostraran las corridas formadas en la primer tabla
corridas = []
corrida_actual = Secuencias[1][2]  # Comienza desde i=1
longitud_corrida = 1 #con el valor 1, porqué representará la longitud actual de la corrida en proceso.

# Conteo de las corridas y sus longitudes
for i in range(2, len(Secuencias)):  # Comienza desde i=2 porque la i=0 es el primer elemento que no contamos e i=1 es donde iniciamos la corrida y por lo tanto debemos comparar a partir del indice 2
    if Secuencias[i][2] == corrida_actual:
        longitud_corrida += 1
    else:
        corridas.append([corrida_actual, longitud_corrida])
        corrida_actual = Secuencias[i][2]
        longitud_corrida = 1

# Agregar los datos a  corridas
corridas.append([corrida_actual, longitud_corrida])

# Enumerar las corridas
for i, corrida in enumerate(corridas):
    corrida[0] = i + 1

# Encabezados de la tabla
headers_corridas = ["Corrida", "Longitud"]
# Imprimir la tabla de corridas usando la función tabulate
tabla_corridas = tabulate(corridas, headers=headers_corridas, tablefmt="grid")
print(tabla_corridas)


# 5. Sustituimos los datos en las fórmulas dadas en el texto inicial
'''Calcularemos los valores para
                   * μ_C_0= (2n-1)/3

                   * σ^2_C_0= (16n-29)/90

                   * Z_0= | (C_0 - μ_C_0)/σ_C_0 |'''

C_0=len(corridas)
print("> C_0=",C_0)
miu_C_0 = round(((2 * n - 1) / 3), 5)
print(f"> μ_{C_0} =", miu_C_0)
sigma_C_0 = round(((16 * n - 29) / 90), 5)
print(f"> σ^2_{C_0} =", sigma_C_0)
Z_0=round(abs((C_0-miu_C_0)/math.sqrt(sigma_C_0)),5)
print("> Z_0=",Z_0)


# Calcular los valores críticos Z_α/2 y -Z_α/2

'''
Como queremos calcular los valores críticos para un intervalo de confianza en la distribución normal estándar
usaremos la función ppf()(Percent Point Function) de la distribución normal estándar para encontrar
el valor crítico correspondiente a un determinado nivel de significancia
Entonces, recordando que para un intervalo de confianza bilateral (de dos colas) con nivel de confianza 1−α
debemos encontrar los valores críticos -Z_α/2 y Z_α/2'''

Z_alpha_2 =round(norm.ppf(1 - alpha/2),2)
neg_Z_alpha_2 =round(-Z_alpha_2,2)

#Valores críticos
print("> Z_α/2:", Z_alpha_2)
print("> -Z_α/2:", neg_Z_alpha_2)

#Resultado, aquí compararemos el valor Z_0 con los -Z_α/2 y Z_α/2 y veremos si aceptamos o no la hip+otesis nula
if neg_Z_alpha_2<=Z_0<=Z_alpha_2:
  print("Como Z_0 cumple la desigualdad:",neg_Z_alpha_2,"≤",Z_0,"≤",Z_alpha_2," aceptamos la hipótesis nula H_0")
  print("H_0:",H_0)
else :
  print("Como Z_0 No cumple la desigualdad:",neg_Z_alpha_2,"≤",Z_0,"≤",Z_alpha_2," aceptamos la hipótesis alternativa H_1")
  print("H_1:",H_1)




```
