# Método del *producto medio*
La mecánica de generación de números pseudoaleatorios de este algoritmo no congruencial es similar a la del algoritmo de cuadrados medios.

**Pasos**


1.   Se requieren de dos *semillas*, x_n y x_n-1, de *p* dígitos
2.   Se multiplican ambas semillas y, del producto, se seleccionan los *p* dígitos del centro, los cuales formarán el primer número pseudoaleatorio al cual llamamos x_n+1
3. Ahora, se mutiplicá está nueva semilla, con la anterior a esta y se repite hasta generar el N ciclo de vida posible

**Nota**: Si no es posible obtener los *p* dígitos del centro del número, agregamos ceros a la izquierda de este.



```python
'''BOSQUEJO para dos semillas sin iteraciones '''

#primero definimos un condicional que determine si los números ingresados son de p digitos:
'''Para mantener la uniformidad y la simplicidad en la implementación del algoritmo,
es común que las semillas tengan la misma longitud'''

longitud = input("Ingrese el número de dígitos a trabajar: ")

x_n1 = input("Ingrese el primer número: ")
long_x_n1 = str(x_n1)  #Lo leemos como cadena de texto para después poder calcular el tamaño de esta cadena, lo que nos devolvera la longitud P

#Con este condicional buscamos que ambas semillas tengan la longitud mencionada al inicio
if len(long_x_n1) != int(longitud):
    print("Recuerde que el número de dígitos debe ser", longitud)
else:
    xn = input("Ingrese el segundo número: ")
    long_x_n = str(xn)

    if len(long_x_n) != int(longitud):
        print("Recuerde que el número de dígitos debe ser", longitud)

#Definimos el codigo para los pasos dados en el algoritmo inicial

multiplicacion = int (x_n1) * int (xn)  #Se multiplican ambas semillas
print("El producto de",x_n1,"*",xn,"es:",multiplicacion) #Lo imprimimos para verificar que la operacioón fue ejecutada correctamente
mult=str(multiplicacion) #Pasamos a cadena de texto esta multiplicación para poder acceder a cada uno de sus dígitos por separado
inicio = (len(mult) - int(longitud)) // 2 # Con esto eliminamos los extremos del resultado para quedarnos con los centrales  al dividro entre 2
                                           #representa el índice de inicio desde donde se deben extraer los dígitos centrales
digitos_centrales=mult[inicio:inicio+int(longitud)] #Se extraen los digitos centrales
print("Los digitos centrales son",digitos_centrales) #comprobamos que se extrajeron correctamente

xn1=int(digitos_centrales)

#Ingresaremos el ejemplo dado en clase:
#p=4
#x_0=2124
#x_1=5113
#el número que debemos de obtener como digitos centrales será: 8600



```

```python
'''Código iterativo'''

longitud = input("Ingrese el número de dígitos a trabajar: ")

x_n1 = input("Ingrese el primer número: ")
long_x_n1 = str(x_n1)

if len(long_x_n1) != int(longitud):
    print("Recuerde que el número de dígitos debe ser", longitud)
else:
    xn = input("Ingrese el segundo número: ")
    long_x_n = str(xn)

    if len(long_x_n) != int(longitud):
        print("Recuerde que el número de dígitos debe ser", longitud)

#Con ayuda del ciclo for, reealizaremos las iteraciones en un rango definido por al menos 20 iteraciones
for i in range(20):
  print("Iteracion ",i)
  multiplicacion = int (x_n1) * int (xn)
  print("El producto de",x_n1,"*",xn,"es:",multiplicacion)
  mult=str(multiplicacion)
  inicio = (len(mult) - int(longitud)) // 2
  digitos_centrales=mult[inicio:inicio+int(longitud)]
  print("Los digitos centrales son",digitos_centrales)
  #Se hace la actualización de las variables para la siguiente iteración

  xn1=int(digitos_centrales) #Nueva semilla igual a los digitos extraidos
  x_n1=xn #Semilla antecesora
  xn=xn1 #semila x_n+1
```

```python
'''Código final con tabulación de resultados'''

from tabulate import tabulate  #Importamos la libreria para realizar la tabla
longitud = input("Ingrese el número de dígitos a trabajar: ")

x_n1 = input("Ingrese el primer número: ")
long_x_n1 = str(x_n1)

if len(long_x_n1) != int(longitud):
    print("Recuerde que el número de dígitos debe ser", longitud)
else:
    xn = input("Ingrese el segundo número: ")
    long_x_n = str(xn)

    if len(long_x_n) != int(longitud):
        print("Recuerde que el número de dígitos debe ser", longitud)

# Inicializar una lista para almacenar los resultados que se imprimirán en la tabla
resultados = []

for i in range(20):
    # Cálculo del producto
    multiplicacion = int(x_n1) * int(xn)

    # Cálculo de los dígitos centrales
    mult_str = str(multiplicacion)
    inicio = (len(mult_str) - int(longitud)) // 2
    digitos_centrales = mult_str[inicio:inicio + int(longitud)]

    # Agregar los resultados a la lista de resultados
    resultados.append((i, x_n1, xn, multiplicacion, digitos_centrales))
    '''
    append() es un método de las listas que se utiliza para agregar
    un elemento al final de la lista. En este caso,
    resultados es una lista, por lo que append() se usa para añadir elementos a esta lista.
    '''
    # Actualizar los valores para la próxima iteración
    xn1 = int(digitos_centrales)
    x_n1 = xn
    xn = xn1

# Encabezados de la tabla
headers = ["Iteración", "x_n1", "xn", "Multiplicación", "Digitos Centrales"]

# Imprimir la tabla usando la función tabulate
tabla = tabulate(resultados, headers=headers, tablefmt="grid")

print(tabla)
```
