**Prueba Kolmogorov-Smirnov**

Se trata de un método no paramétrico sencillo para probar si existe una diferencia significativa entre una distribución de frecuencias observada y una distribución de frecuencias teórica.
Es por consiguiente otra *medida de bondad de ajuste* de una distribución de frecuencia teórica.

El *estadístico D*, es particularmente útil para juzgar qué tan cerca está la distribución de FO de la distribución de FE. Puesto que éste depende del tamaño de muestra n, pero es independiente de la distribución de frecuencias esperadas. Es un estadístico de distribución libre.

*Pasos*
1. Definir la hipótesis nula H_0 y la hipótesis alternativa H_1
2. Definir el nivel de significancia (alpha) que será igual a:

       α = 1-Nivel de aceptación

3. Ordenar nuestra lista de datos de menor a mayor
4. Calculamos el estadístico de prueba D
      
         D=Max{D^+,D^-}
Donde
         D^+ = Max  |(i/n) - x_i|
              1<i<n

         D^- = Max |x_i - (i-1)/n|
              1<i<n

       *Entendiendo que x_i son los números ta ordenados y (i/n) Es la distribución de frecuencia teórica.* y (i-1)/n es el lado izquierdo

5. Compraramos el valor critico
d_α,n , el cual es un valor dado en la *Tabla de Lilliefors (Test k-s)* con el estadístico D
       -Se acepta H_0 si D<d_α,n
       -Caso contrario, no se acepta
